## a) Importing Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

file_path = "cardio_good_fitness.csv"
df = pd.read_csv(file_path)
```



## b) Current Data
From the above 2 cells, we can observe that the data is not ready for analysis and requires some preparation. 

Currently, there are a few issues with the data given that make it hard to work with. 

Firstly there are multiple null values and they are not in the right format of <NA>.

Secondly, the data given/imported is not in the right format. 

Thirdly, there are some columns with different naming conventions (e.g 'Single' and 'S')

Hence preparation is needed to solve these issues to ensure it is good for analysing.


## c) Removing Null Values

```python
df.head
```

```python
#Filling in the null values with "NA" to be changed later on.

#Replacing the null values/na values with valid NA values (<NA>).
df.replace(['na', 'NA', '<NA>'], pd.NA, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.head
```

Now it shows that all of the NA values have been removed.

## d) Changing Data Types
To resolve the 2nd issue- Changing the Data Types to the appropriate data types accordingly

**Column Name**      Before Data Type -> After Data Type

**Product**         object -> category

**Branch**           object -> category

**Age**              object -> int32

**Gender**           object -> category

**Education**        object -> int32

**MaritalStatus**    object -> category

**Usage**            object -> int32

**Fitness**          object -> int32

**Income**           object -> int32

**Miles**            object -> int32


### Marital Status formats were not standardised, some had "Single" and some has "S". Hence I had to replace the values to make it easier to work with

```python
# Ensure 'MaritalStatus' is a string before replacing values
df['MaritalStatus'] = df['MaritalStatus'].astype(str).str.strip()
#Ensure naming convention is standardised
df['MaritalStatus'] = df['MaritalStatus'].replace({'S': 'Single', 'P': 'Partnered'})
```

### Changing data types of these columns into "int32"

```python
num_cols = ['Age', 'Education', 'Usage', 'Fitness', 'Income', 'Miles']
for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')  # Convert to numeric first
    df[col] = df[col].fillna(0).astype(int)
```

### Changing data types of these columns into "category"

```python
cat_cols = ['Product', 'Branch', 'Gender', 'MaritalStatus']
for col in cat_cols:
    df[col] = df[col].astype('category')
```

```python
print(df.dtypes)
```

From this cell, you can observe that the data types has successfully been changed into the appropriate ones

### Standardize Text Data

```python

#Some values might have inconsistent capitalization or leading spaces
df['Product'] = df['Product'].str.upper()  # Ensures consistency
df['Branch'] = df['Branch'].str.strip().str.title()
df['Gender'] = df['Gender'].str.strip().str.capitalize()  # Ensures "Male" not "male"
```

## Visualisation 10

```python
fig, ax1 = plt.subplots(figsize=(8, 5))

sns.barplot(
    y='Product',  
    x='Usage', 
    data=df, 
    estimator=lambda x: x.mean(),  
    hue='Product', 
    palette="viridis", 
    errorbar=None,  
    legend=False, 
    ax=ax1
)

ax2 = ax1.twiny()

miles_avg = df.groupby('Product', observed=False)['Miles'].mean().sort_values()
ax2.plot(miles_avg.values, miles_avg.index, marker='o', color='red', linestyle='--', linewidth=2, label="Avg Miles")

for p in ax1.patches:
    ax1.annotate(f"{p.get_width():.2f}", 
                 (p.get_width() + 0.1, p.get_y() + p.get_height() / 2),
                 ha='left', va='center', fontsize=12, fontweight='bold', color='black')

for i, (miles, product) in enumerate(zip(miles_avg.values, miles_avg.index)):
    ax2.annotate(f"{miles:.2f}", 
                 (miles + 5, i), 
                 ha='left', va='center', fontsize=12, fontweight='bold', color='red')

ax1.set_xlabel("Average Weekly Usage (Times per Week)", fontsize=12)
ax1.set_ylabel("Treadmill Model", fontsize=12)
ax2.set_xlabel("Average Miles per Week", fontsize=12, color='red')
ax1.set_title("Average Usage & Miles per Week vs. Treadmill Model", fontsize=14, fontweight='bold')

ax1.set_xlim(0, df['Usage'].max() + 1)  
ax2.set_xlim(0, miles_avg.max() + 10)  

ax1.grid(False)
ax2.grid(False)

plt.show()
```
